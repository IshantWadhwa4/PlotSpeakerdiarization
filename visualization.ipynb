```python
def getEnd(pointer,label):
    count = 1
    if pointer > len(label)-1:
        return count,pointer
    while pointer < len(label)-1:
        #print(pointer,label[pointer])
        if label[pointer][0] == label[pointer+1][0]:
            count = count+1
            pointer +=1
        else:
            break
        
    return count,pointer
            
        
    
```
